## 4. Bullet Lists and Numbered Lists

Use bullet lists for unordered items:


```markdown

3. Vectorize the text.
4. Train the model.
5. Evaluate the model.
```

## 7. Markdown in Jupyter Notebooks

A good notebook should not be only code.

It should tell the story of the experiment:

1. What is the question?
2. What data are used?
3. What method is used?
4. What result was obtained?
5. What does the result mean?
6. What are the limitations?
7. Was GenAI used? If yes, how?

A notebook with working code but no explanation is incomplete.

## 9. GenAI Use Log in Markdown

In this course, GenAI use is not automatically prohibited. However, it must be visible, bounded, auditable, and pedagogically meaningful.

Example:

```markdown

I did not ask it to generate the final interpretation. 
I revised the explanation after checking the model results myself.
```

## 11. Minimal Markdown Checklist

Before submitting a Markdown-based notebook or report, check the following:

- The notebook has a clear title.
- Sections are organized with headings.
- Code is explained using Markdown cells.
- Important filenames, column names, and variables use inline code.
- Results are summarized in words, not only printed as output.
- Tables are used when comparing models or metrics.
- Limitations are stated honestly.
- GenAI use, if any, is documented.
